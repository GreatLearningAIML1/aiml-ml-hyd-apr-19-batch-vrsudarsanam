```python
df['StartYear']=df['Debut'].str[0:4].astype(int)
DebutSeries= df[(df['StartYear']>=1930) & (df['StartYear']<=1980)]
DebutSeries[['Team','StartYear']].sort_values('StartYear')                                     
#DebutSeries.sort('StartYear')                                       
```

# 4. Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. 

## Using the same function, find the team which has maximum and minimum goal difference.

```python
def Goal_Diff_Count(Ds):
   
    Ds.index=Ds['Team']
    Ds[['GoalsFor', 'GoalsAgainst']] = Ds[['GoalsFor', 'GoalsAgainst']].apply(pd.to_numeric)
   
    
    Diff = Ds['GoalsFor']-Ds['GoalsAgainst']
    Diff.columns='Goal Diff'
    return Diff

#print(Goal_Diff_Count(df))

print(Goal_Diff_Count(df).sort_values())


    
```

```python
print("Min Point difference by a team")
print(Goal_Diff_Count(df).sort_values().head(1))
print("Max Point difference by a team")
print(Goal_Diff_Count(df).sort_values().tail(1))
```
