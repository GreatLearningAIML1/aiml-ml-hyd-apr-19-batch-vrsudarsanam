##### 2. It is asked that whether the new scheme has significantly raised the output, it is an example of the one-tailed t-test.


Let 
Mean1 = Average sums assured by salesperson BEFORE changeover. Mean2 = Average sums assured by salesperson AFTER changeover.

NULL Hypothesis: Mean1 = Mean2  ; Mean2 – Mean1  = 0

Alternative Hypothesis: Mean1 < Mean2   ; Mean2 – Mean1  > 0 ; true difference of means is greater than zero.

Since population standard deviation is unknown, paired sample t-test will be used.

#### Using user defined function to calculate the p-value and t_stat values 

```python
from math import sqrt
from numpy.random import seed
from numpy.random import randn
from numpy import mean
from scipy.stats import t

def dependent_ttest(data1, data2, alpha):
    # calculate means
    mean1, mean2 = mean(data1), mean(data2)    
    # number of paired samples
    n = len(data1)
    # sum squared difference between observations
    d1 = sum([(data1[i]-data2[i])**2 for i in range(n)])
    # sum difference between observations
    d2 = sum([data1[i]-data2[i] for i in range(n)])
    # standard deviation of the difference between means
    sd = sqrt((d1 - (d2**2 / n)) / (n - 1))
    # standard error of the difference between the means
    sed = sd / sqrt(n)
    # calculate the t statistic
    t_stat = (mean1 - mean2) / sed
    # degrees of freedom
    degreesoffreedom = n - 1
    # calculate the critical value
    cv = t.ppf(1.0 - alpha, degreesoffreedom)
    # calculate the p-value
    p = (1.0 - t.cdf(abs(t_stat), degreesoffreedom)) * 2.0
    # return everything
    return t_stat, degreesoffreedom, cv, p

t_stat, degreesoffreedom, cv, p = dependent_ttest(df.iloc[:,2], df.iloc[:,1], 0.05)
```

```python
print('t=%.3f, degreesoffreedom=%d, cv=%.3f, p=%.3f' % (t_stat, degreesoffreedom, cv, p))
```

#### Cross checking the values obtained above with the scipy.stats ttest_rel function

```python
from scipy.stats import ttest_rel
t_stat, p_stat = ttest_rel(df.iloc[:,2], df.iloc[:,1])
print('t=%.3f, p_stat=%.3f'% (t_stat, p_stat))
```
